```python
#Sort data by Ticker and Date
df = df.sort_values(['Ticker', 'Date']).reset_index(drop=True)
```

# ***Business Insights***

```python
#Daily returns
df['daily_return']=df.groupby('Ticker')['Close'].pct_change()
```

```python
#Moving average
df['MA_7'] = df.groupby('Ticker')['Close'].transform(lambda x: x.rolling(7).mean())
df['MA_30'] = df.groupby('Ticker')['Close'].transform(lambda x: x.rolling(30).mean())
```

```python
#Volatility
df['Volatility_7'] = df.groupby('Ticker')['daily_return'].transform(lambda x: x.rolling(7).std())
df['Volatility_30'] = df.groupby('Ticker')['daily_return'].transform(lambda x: x.rolling(30).std())
```

```python
#Normalised price
df['Normalized_Price'] = df.groupby('Ticker')['Close'].transform(lambda x: x / x.iloc[0] * 100)
```

```python
#Cumulative returns
df['Cumulative_Return'] = df.groupby('Ticker')['daily_return'].transform(lambda x: (1 + x).cumprod())
```

```python
#Correlation
returns = df.pivot(index='Date', columns='Ticker', values='daily_return')
returns = returns.dropna()
corr_matrix = returns.corr()
```

```python
#Drawdown Analysis
def drawdown(series):
    cum = (1 + series).cumprod()
    peak = cum.cummax()
    dd = (cum - peak) / peak
    return dd

df['Drawdown'] = df.groupby('Ticker')['daily_return'].transform(drawdown)

plt.figure(figsize=(10,6))
for t in df['Ticker'].unique():
    plt.plot(df[df['Ticker']==t]['Date'], df[df['Ticker']==t]['Drawdown'], label=t)
plt.title("Max Drawdown")
plt.legend(); plt.show()
```

```python
#Rolling sharpe ratio
df['Rolling_Sharpe_20'] = df.groupby('Ticker')['daily_return'].transform(
    lambda x: x.rolling(window=20).mean() / x.rolling(window=20).std()
)

plt.figure(figsize=(10,6))
for ticker in df['Ticker'].unique():
    subset = df[df['Ticker'] == ticker]
    plt.plot(subset['Date'], subset['Rolling_Sharpe_20'], label=ticker)

plt.axhline(0, color='black', linewidth=1, linestyle='--')  # zero threshold
plt.title("Rolling 20-Day Sharpe Ratio Comparison")
plt.xlabel("Date")
plt.ylabel("Sharpe Ratio")
plt.legend()
plt.tight_layout()
plt.show()
```
